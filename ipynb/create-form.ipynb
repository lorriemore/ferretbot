# eecord book club form creator
the eecord book club forms are always the same. this should be able to, given a list of seconded books (how that's worked out is a later step) and a date range (also to be worked out) generate a google form following the format required.

there'll then also need to be a results processor that'll process the results a specified number of days after the form is sent out, and it'll need to be able to announce the form going live and give a reminder in the book club thread a day before it closes.

we probably also want it to be able to create discord events.

then we need it to be able to interact with people, probably in the following ways:
- adding a book to the suggestions list
- seconding a book (check if user is in VC and the event is live at the time?)
- announcing new forms, next event, event reminders, etc.
- listing current suggestions
- specifying date range for next vote
